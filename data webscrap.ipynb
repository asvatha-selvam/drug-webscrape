```python
import requests
from bs4 import BeautifulSoup
import csv

def extract_comments_from_page(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the specific div with class "contentBox"
        content_box = soup.find("div", class_="contentBox")

        # Check if the contentBox is found
        if content_box:
            # Extract user comments based on the HTML structure
            user_comments = content_box.find_all("div", class_="ddc-comment")

            # Create a CSV file and write header
            with open("nivolumab_reviews.csv", "a", newline="", encoding="utf-8") as csvfile:
                csv_writer = csv.writer(csvfile)

                # Iterate through user comments and extract relevant data
                for i, comment in enumerate(user_comments, start=1):
                    # Extract the date of the review
                    date_element = comment.find("ul", class_="ddc-comment-header").find("li", class_=None)
                    date = date_element.text.strip() if date_element else "Date not available"

                    # Extract the condition for the review
                    condition_element = comment.find("p").find("b")
                    condition = condition_element.text.strip() if condition_element else "Condition not available"

                    # Extract the review text
                    review_text_element = comment.find("p").find_all(text=True, recursive=False)
                    review_text = ''.join(review_text_element).strip() if review_text_element else "Review text not available"

                    # Write the extracted data to the CSV file
                    csv_writer.writerow([date, condition, review_text])

                    print(f"Review {i} Date: {date}")
                    print(f"Condition: {condition}")
                    print(f"Review Text: {review_text}")
                    print("-" * 50)

            print(f"Data from page '{url}' has been saved to 'nivolumab_reviews.csv'.")
        else:
            print(f"ContentBox not found on the page: {url}")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

#Base URL for the comments
base_url = "https://www.drugs.com/comments/nivolumab/"

#Iterate through multiple pages (adjust the range based on your needs)
for page_number in range(1, 4):  # Change the range as needed
    page_url = f"{base_url}?page={page_number}"
    extract_comments_from_page(page_url)
```
